start_date and end_date columns don't have the right format. Let's fix it.

```python
from datetime import datetime
# The function below transforms an integer date format to a datetime format easier to manipulate
def int2date(int_date):
    string = str(int_date)
    date = datetime(year=int(string[0:4]), month=int(string[4:6]), day=int(string[6:8]))
    return date
```

```python
#applymap is a powerful method that applies a function on a Dataframe columns elementwise
calendar[['start_date','end_date']] = calendar[['start_date','end_date']].applymap(int2date)
```

Waterfall train schedule above gives time details around 14:23 which is not the case for the Banana Bus. The frequencies table below gives information about the Bus service start tiime and headway. We can draw a space-time schedule for both services to better locate the observer in the middle of the beach at 14:23.

Assumptions: 


<li> Buses and trains operating on both lines change trip headsign when arriving to terminus.

<li>Departure and arrival times might be not equal in reality because for operational efficiency buses and trains are generally used on the 2 way trips. For instance, the tain used for Waterfall_0004 trip_id might be the one used for Waterfall_0001. The tain arrives to Island City at 14:15 waits 5 minutes and is used again on trip_id Waterfall_0002. Considering this logic, i suspect that arrival_time and departure_time could have wrong values.
The following space-time diagrams show both train and bus services working on a cyclic way.

To find the number of Buses/Trains on a service we use the followwing formula:

$$
N_{Buses/Trains} = \frac{cycle_{duration}}{headway}
$$

$$
N_{Buses} = \frac{40min}{10min} = 4Buses
$$


$$
N_{Trains} = \frac{45min}{20min} = 2,25trains
$$

We round it to the upper integer to get:
$$
N_{Trains} = 3Trains
$$

```python
import datetime as dt

#time reference equal to bus service starting time according to the frequencies table 14:05
ref_bus = dt.datetime(2020,1,1,14,5)
#Time deltas in minutes for a given bus cycle
deltas = [5,2,8,5,8,2,5,5,5,2,8,5,8,2,5,5]


bus_1_times = [ref_bus]

for delta in deltas:
    bus_1_times.append(bus_1_times[-1]+dt.timedelta(minutes=delta))
    
bus_2_times=[]
bus_3_times=[]
bus_4_times=[]

#shift bus_1_times back with 1 and 2 time long headways
for time in bus_1_times:
    bus_2_times.append(time + dt.timedelta(minutes=-10))
    bus_3_times.append(time + dt.timedelta(minutes=-20))
    bus_4_times.append(time + dt.timedelta(minutes=-30))

#The observer coordinates
x_observer = [dt.datetime(2020,1,1,14,23,0)]
y_observer = [0]


import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig0, ax0 = plt.subplots(figsize=(10,8))

#Distances traveled between 2 consecutive time steps 
distances = [0,1,1,2,2,1,1,0,0,1,1,2,2,1,1,0,0]

#Selecting only interest time around 14:23
ax0.plot(bus_1_times[0:8],distances[0:8])
ax0.plot(bus_2_times[2:9],distances[2:9])
ax0.plot(bus_3_times[4:12],distances[4:12])
ax0.plot(bus_4_times[5:13],distances[5:13])
ax0.scatter(x_observer, y_observer)

labels0 = [item.get_text() for item in ax0.get_yticklabels()]
labels0[1] = 'The Beach (South) Rail Station'
labels0[5] = 'Bamboo Crossroads'
labels0[9] = 'TIsland City Central Station'

ax0.set_yticklabels(labels0)

ax0.legend(['BananaBus','BananaBus','BananaBus','BananaBus'], loc="center right")

plt.title('Fig -1- Banana Buses space-time diagram schedule')

ax0.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
```

```python
#time reference equal to train service starting time according to the frequencies table 14:00
ref_rail = dt.datetime(2020,1,1,14,0,0)
train_1_times = [ref_rail]

#Time deltas(steps) in minutes for a given train cycle
deltas_rail = [15,5,15,10,15,5,15,10]

#Distances traveled between 2 consecutive time steps
train_distances = [0,1,1,0,0,1,1,0,0]

for delta in deltas_rail:
    train_1_times.append(train_1_times[-1] + dt.timedelta(minutes = delta))

train_2_times = []
train_3_times = []

#creating train_2 and train_3 times by shifting back Train_1 time by 1 and 2 times headways
for time in train_1_times:
    train_2_times.append(time + dt.timedelta(minutes=-15))
    train_3_times.append(time + dt.timedelta(minutes=-30))

fig, ax = plt.subplots(figsize=(10,8))

#Selecting only interest time around 14:23
ax.plot(train_1_times[0:4],train_distances[0:4])
ax.plot(train_2_times[1:5],train_distances[1:5])
ax.plot(train_3_times[3:6],train_distances[3:6])
ax.scatter(x_observer, y_observer)


labels = [item.get_text() for item in ax.get_yticklabels()]
labels[1] = 'The Beach (North) Rail Station'
labels[6] = 'Island City Central Station'

ax.set_yticklabels(labels)

ax.legend(['Waterfall','Waterfall','Waterfall'], loc="center right")
plt.title('Fig -2- Waterfall Railway space-time diagram schedule')

ax.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M"))
```

fig-1- and Fig-2- show how Banana Buses and Waterfall trains operate, respectively. The dot point in both figures at 14:23 show that in both cases the observer is seeing a Banana bus and a Waterfall train that arrived at 14:20 and are stopped waiting for the next departure time. So given the assupmtions that we made and the figures, the observer is most likely to see a train and a bus at the same time.
